```python
import os 
import json 
from collections import Counter
```

```python
lines = []
with open(os.path.join('..', 'data', 'arxiv_data.jsonl'), 'r') as file:
    for line in file: 
        lines.append(json.loads(line))
```

Preview one entry in the arXiv

```python
lines[0]
```

See distribution of categories

```python
categories = Counter()

for line in lines: 
    categories.update(
        [line.get('categories')]
    )
```

```python
categories.most_common(15)
```

Q: Are all entries unique? 

A: Not all entires are unique, some publication ids have multiple records

```python
pub_ids = set()

_ = [pub_ids.add(line.get('id')) for line in lines]
```

```python
len(pub_ids) == len(lines)
```

```python
len(pub_ids), len(lines)
```

Q: Is the updated field ever used? 

A: Yes, about a third of entries get updated. Some entries are updated more than once. 

```python
# unique ids that are updated 
updated_pubs_unique = set()

# all instances of updates
update_count = 0

for line in lines: 
    if line.get('updated') is not None: 
        pub_id = line.get('id')
        updated_pubs_unique.add(pub_id)
        update_count += 1
```

```python
len(updated_pubs_unique), update_count
```

```python
# percent that get updated
len(updated_pubs_unique)/len(lines)
```
